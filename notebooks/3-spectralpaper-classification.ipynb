  <div class="card">
    <div class="header"><span class="dot" style="background:var(--card1)"></span><h3>LDA <span class="badge gradient1">LD</span></h3></div>
    <p>Linear boundaries from shared-covariance Gaussians.</p>
    <div class="eq">\( \delta_c(x)=x^\top\Sigma^{-1}\mu_c-\tfrac12\mu_c^\top\Sigma^{-1}\mu_c+\ln\pi_c \)</div>
  </div>

  <div class="card">
    <div class="header"><span class="dot" style="background:var(--card2)"></span><h3>Gaussian Naive Bayes <span class="badge gradient2">KNB, GNB</span></h3></div>
    <p>Independence across features; fast generative classifier.</p>
    <div class="eq">\( P(y=c\mid x)\propto \pi_c \prod_j \mathcal{N}(x_j;\mu_{cj},\sigma_{cj}^2) \)</div>
  </div>

  <div class="card">
    <div class="header"><span class="dot" style="background:var(--card3)"></span><h3>Decision Tree <span class="badge gradient3">FT</span></h3></div>
    <p>Greedy splits to reduce impurity.</p>
    <div class="eq">\( \Delta = G_{\text{parent}}-\tfrac{w_L}{w}G_L-\tfrac{w_R}{w}G_R \quad (G=1-\sum_c p_c^2) \)</div>
  </div>

  <div class="card">
    <div class="header"><span class="dot" style="background:var(--card4)"></span><h3>k‚ÄëNN <span class="badge gradient4">FKNN, SKNN</span></h3></div>
    <p>Vote of neighbors (distance‚Äëweighted in SKNN).</p>
    <div class="eq">\( w_i=\frac{1}{\lVert x-x_{(i)}\rVert+\varepsilon},\quad \hat y=\arg\max_c \sum_{i\in\mathcal N_k} w_i\,\mathbf1[y_i=c] \)</div>
    <p class="hint">Scale features; choose k by CV.</p>
  </div>

  <div class="card">
    <div class="header"><span class="dot" style="background:var(--card5)"></span><h3>Bagging (Trees) <span class="badge gradient5">BT</span></h3></div>
    <p>Bootstrap many trees; average votes ‚Üí lower variance.</p>
    <div class="eq">\( \hat y=\arg\max_c \sum_{b=1}^B \mathbf1[h_b(x)=c] \)</div>
  </div>

  <div class="card">
    <div class="header"><span class="dot" style="background:var(--card6)"></span><h3>SVM Family <span class="badge gradient6">SVM_*</span></h3></div>
    <p>Max‚Äëmargin classifier; kernels add nonlinearity.</p>
    <div class="eq">Linear: \( f(x)=w^\top x+b \) (soft‚Äëmargin \(C\)).</div>

# üìä Understanding Multi-Class Classification Metrics  

When we evaluate a classification model (e.g., for **Normal**, **Asthma**, **Pneumonia** sounds), we rely on metrics derived from the **Confusion Matrix**.

---

## üîπ Confusion Matrix Terms

For each class (one-vs-all approach):

- **TP (True Positive):**  
  Model correctly predicts samples of this class as belonging to the class.  

- **TN (True Negative):**  
  Model correctly predicts samples of other classes as *not* belonging to this class.  

- **FP (False Positive):**  
  Model incorrectly assigns other class samples into this class.  

- **FN (False Negative):**  
  Model fails to recognize samples of this class (misclassified as another class).  

---

<div style="background: linear-gradient(135deg, #e8f4fd, #f0f7ff); 
            padding: 20px; 
            border-radius: 12px; 
            border: 1px solid #cce0ff;
            font-family: 'Segoe UI', sans-serif;">

<h2 style="color:#003366;">üî∑ Formulas (Per Class)</h2>

- **Accuracy:**  
\[
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
\]

- **Precision (Positive Predictive Value):**  
\[
Precision = \frac{TP}{TP + FP}
\]  
<span style="color:#0073e6;">üîπ Of all predicted positives, how many are truly correct?</span>

- **Recall (Sensitivity / True Positive Rate):**  
\[
Recall = \frac{TP}{TP + FN}
\]  
<span style="color:#0073e6;">üîπ Of all actual positives, how many did the model correctly identify?</span>

- **F1-Score (Balance of Precision & Recall):**  
\[
F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}
\]

</div>


---

## üîπ Macro & Micro Averaging (Multi-Class)

Since we have **3 classes (Normal, Asthma, Pneumonia)**, metrics can be averaged in two ways:

- **Macro Average:**  
  Compute Precision/Recall/F1 for each class separately, then take the **unweighted mean**.  
  *All classes treated equally, regardless of size.*

- **Micro Average:**  
  Aggregate **all TP, FP, FN across classes**, then compute metrics.  
  *Better when classes are imbalanced.*

---

‚úÖ These metrics ensure a **comprehensive evaluation** of our classifiers, beyond raw accuracy, especially important in medical diagnosis where **false negatives (FN)** can be critical.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    confusion_matrix, classification_report, accuracy_score, 
    precision_score, recall_score, f1_score
)
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC

# -----------------------------
# LOAD DATA
# -----------------------------
df = pd.read_excel("/kaggle/input/spectral-paper/lung_sound_features.xlsx")

# Drop only the 'Class' column since 'File' doesn't exist
X = df.drop(columns=["Class"]).values
y = df["Class"].values


# -----------------------------
# TRAIN-TEST SPLIT
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

print(f"Train size: {len(y_train)}, Test size: {len(y_test)}")

# -----------------------------
# MODEL DEFINITIONS
# -----------------------------
models = {
    "LD": LinearDiscriminantAnalysis(),
    "KNB": GaussianNB(),  # Kernel NB approximation with Gaussian NB
    "FT": DecisionTreeClassifier(max_depth=None, random_state=42),  # Fine Tree
    "GNB": GaussianNB(),
    "FKNN": KNeighborsClassifier(n_neighbors=1),  # Fine KNN
    "BT": BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=50, random_state=42),
    "SKNN": KNeighborsClassifier(n_neighbors=5, weights='distance'),  # Subspace KNN approx
#    "SVM_linear": SVC(kernel="linear", probability=True, random_state=42),
#    "SVM_quadratic": SVC(kernel="poly", degree=2, probability=True, random_state=42),
#    "SVM_cubic": SVC(kernel="poly", degree=3, probability=True, random_state=42),
#    "SVM_fg": SVC(kernel="rbf", gamma=5, probability=True, random_state=42),   # Fine Gaussian
#    "SVM_mg": SVC(kernel="rbf", gamma=0.5, probability=True, random_state=42), # Medium Gaussian
#    "SVM_cg": SVC(kernel="rbf", gamma=0.05, probability=True, random_state=42) # Coarse Gaussian
}

# -----------------------------
# EVALUATION FUNCTION
# -----------------------------
def evaluate_model(name, model, X_train, X_test, y_train, y_test, cmap="Blues"):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="macro")
    rec = recall_score(y_test, y_pred, average="macro")
    f1 = f1_score(y_test, y_pred, average="macro")

    # Type I & II errors
    cm = confusion_matrix(y_test, y_pred, labels=np.unique(y))
    FP = cm.sum(axis=0) - np.diag(cm)  # false positives
    FN = cm.sum(axis=1) - np.diag(cm)  # false negatives
    TP = np.diag(cm)
    TN = cm.sum() - (FP + FN + TP)

    type1 = FP / (FP + TN + 1e-12)  # False Positive Rate
    type2 = FN / (FN + TP + 1e-12)  # False Negative Rate

    print(f"\nüîπ {name} Results üîπ")
    print(f"Accuracy: {acc:.3f}, Precision: {prec:.3f}, Recall: {rec:.3f}, F1: {f1:.3f}")
    print("Type I Error (per class):", np.round(type1,3))
    print("Type II Error (per class):", np.round(type2,3))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

    # Confusion Matrix Plot
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt="d", cmap=cmap, xticklabels=np.unique(y), yticklabels=np.unique(y))
    plt.title(f"{name} Confusion Matrix", fontsize=14, weight="bold")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

# -----------------------------
# RUN ALL MODELS
# -----------------------------
import os
import joblib  # for saving models

# -----------------------------
# CREATE OUTPUT DIRECTORY
# -----------------------------
os.makedirs("/kaggle/working/output", exist_ok=True)

# -----------------------------
# RUN ALL MODELS & SAVE
# -----------------------------
cmaps = ["Blues","Reds","Greens","Oranges","Purples","BuPu","YlGnBu",
         "cividis","coolwarm","plasma","mako","rocket","Spectral"]

for (i, (name, model)) in enumerate(models.items()):
    evaluate_model(name, model, X_train, X_test, y_train, y_test, cmap=cmaps[i % len(cmaps)])
    
    # save trained model
    save_path = f"output/{name}_model.pkl"
    joblib.dump(model, save_path)
    print(f"‚úÖ Saved {name} model to {save_path}")
```

```python
# -----------------------------
# COLLECT RESULTS FOR COMPARISON
# -----------------------------
results = []

for (name, model) in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average="macro")
    prec = precision_score(y_test, y_pred, average="macro")
    rec = recall_score(y_test, y_pred, average="macro")

    results.append({
        "Model": name,
        "Accuracy": acc,
        "F1": f1,
        "Precision": prec,
        "Recall": rec
    })

results_df = pd.DataFrame(results)
display(results_df)

# -----------------------------
# BAR PLOTS FOR COMPARISON
# -----------------------------
plt.figure(figsize=(12,6))
sns.barplot(x="Model", y="Accuracy", data=results_df, palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.title("Classifier Accuracy Comparison", fontsize=14, weight="bold")
plt.ylabel("Accuracy")
plt.show()

plt.figure(figsize=(12,6))
sns.barplot(x="Model", y="F1", data=results_df, palette="magma")
plt.xticks(rotation=45, ha="right")
plt.title("Classifier F1-Score Comparison", fontsize=14, weight="bold")
plt.ylabel("F1-score")
plt.show()
```

  <h2 style="margin-top:0; text-align:center; font-size:24px;">üßÆ Statistical Significance with McNemar‚Äôs Test</h2>

  <p style="font-size:15px; line-height:1.6;">
    High accuracy alone doesn‚Äôt guarantee that one model is <b>truly better</b> than another.  
    To test whether the difference in performance is <b>statistically significant</b> (not random chance), 
    we use <b>McNemar‚Äôs Test</b>.
  </p>

  <hr style="border: 1px solid rgba(255,255,255,0.3);">

  <h3 style="margin-bottom:6px;">üìå How It Works</h3>
  <ul style="font-size:14px; line-height:1.6;">
    <li>We compare each classifier directly against the <b>best SVM variant</b> (Medium Gaussian in this case).</li>
    <li>Counts:
      <ul>
        <li>üîπ <b>b</b>: Cases competitor is correct, but SVM is wrong</li>
        <li>üîπ <b>c</b>: Cases SVM is correct, but competitor is wrong</li>
      </ul>
    </li>
    <li>If <b>b ‚âà c</b>, both models perform similarly.</li>
    <li>If <b>b and c differ significantly</b>, one model is <b>statistically better</b>.</li>
  </ul>

  <p style="font-size:14px; line-height:1.6;">
    ‚úÖ Interpretation:
    <br>‚Ä¢ <b>p-value &lt; 0.05</b> ‚Üí Significant difference (SVM truly outperforms competitor).  
    <br>‚Ä¢ <b>p-value ‚â• 0.05</b> ‚Üí No significant difference (performances too close).  
  </p>

  <hr style="border: 1px solid rgba(255,255,255,0.3);">

  <p style="font-size:14px; line-height:1.6;">
    üîé This statistical test gives us confidence that our <b>SVM results are reliable</b> 
    and not just due to randomness in the dataset.
  </p>
</div>


```python
from statsmodels.stats.contingency_tables import mcnemar

# Pick the best model 
best_model = models["FT"]
best_model.fit(X_train, y_train)
y_best = best_model.predict(X_test)

print("üîé Statistical Significance Testing (McNemar‚Äôs Test vs SVM)")

for name, model in models.items():
    if name == "FT":  # skip self-comparison
        continue
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Build contingency table
    # rows: correct/incorrect by SVM
    # cols: correct/incorrect by competitor
    svm_correct = (y_best == y_test)
    other_correct = (y_pred == y_test)

    b = np.sum(~svm_correct & other_correct)  # competitor correct, SVM wrong
    c = np.sum(svm_correct & ~other_correct)  # SVM correct, competitor wrong

    table = [[0, b], [c, 0]]

    result = mcnemar(table, exact=True)  # exact binomial version
    print(f"\n{name} vs SVM_mg")
    print(f"b={b}, c={c}, McNemar p-value={result.pvalue:.5f}")
    if result.pvalue < 0.05:
        print("‚û°Ô∏è Significant difference at 95% confidence")
    else:
        print("‚û°Ô∏è No significant difference (performances too close)")
```
